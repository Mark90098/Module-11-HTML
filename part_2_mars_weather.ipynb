```python
#I've converted the query results into a list more than once so I made a custom function to do it for me
#row_element is the input, the long query results.
#column_tag is HTML element it searches for. Default is 'td', used for table body. Table columns needs 'th'
def row_to_columns(row_element,column_tag='td'):
    one_row = []
    row_data = row_element.find_all(column_tag)
    for column in row_data:
        one_row.append(column.text)
    return one_row
```

```python
#grab column titles and make it into a list. coulumn_tag is not 'td' default, it is 'th' because it's the table header
header_columns = row_to_columns(header_row,'th')
header_columns
```

```python
#same code as 3 cells up, but using my function
biglist = []

for row in row_body:
    one_row = row_to_columns(row)
    biglist.append(one_row)
biglist
```

```python
df = pd.DataFrame(biglist,columns=header_columns)
df
```

```python
#Create earth days varible from date column with for loop.

row = df.iloc[:,1]
earth_days=[]
for row in row:
    earth_days.append((row - df.iloc[0,1])/datetime.timedelta(days=1)) #need to divide by timedelta to get integer to plot on scatterplot
   # print(row)
#earth_days
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
#make scatterplot. looks like about 750 earth days = 1 martian year
plt.scatter(earth_days,df['min_temp'])
plt.xlabel = ('# of Earth days')
plt.ylabel = ("Min Temp")
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
